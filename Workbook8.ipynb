```python
# â„– 1

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

x = np.array([[5, 3], 
             [10, 15], 
             [15, 12], 
             [24, 10], 
             [30, 45], 
             [85, 70], 
             [71, 80], 
             [60, 78], 
             [55, 52], 
             [80, 91]])
print(x.shape)

kmeans = KMeans(n_clusters=5, random_state=0)
clusters = kmeans.fit_predict(x.data)
print(kmeans.cluster_centers_.shape)

fig, ax = plt.subplots(1, 5, figsize=(8, 3))
centers = kmeans.cluster_centers_.reshape(5, 2, 1)

for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center, interpolation='nearest', cmap=plt.cm.binary)
```

```python
# â„– 3

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering

dataset = sns.load_dataset('iris')
iris = dataset.loc[:, 'sepal_length' : 'petal_width']
display(iris.head())
print(iris.shape)

data = iris.iloc[:, 2:4].values
plt.figure(figsize=(15, 12), dpi=180)
plt.title("Dendrogram")
dend = shc.dendrogram(shc.linkage(data, method='ward'))

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
print(cluster.fit_predict(data))

plt.figure(figsize=(9, 6))
plt.scatter(data[:, 0], data[:, 1], c=cluster.labels_, cmap='rainbow')
plt.show()
```
