```python
# № 4.3.1

import numpy as np
import matplotlib.pyplot as plt

array = np.random.rand(100)
print("Array:", array)
mean = np.mean(array)
print("Mean value =", mean)
median = np.median(array)
print("Median value =", median)
plt.scatter(np.arange(0, 100), array)
plt.grid()
plt.show()
```

```python
# № 4.3.2

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11)
y = np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x ** 2)) / np.abs(1 - np.sin(x) ** 3) + np.log(np.abs(2 * x))
print("X:", x)
print("Y:", y)
plt.grid()
plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.scatter(x[0:5], y[0:5])
plt.show()
```

```python
# № 4.3.3

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 11)
y = np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x + 1))))
print("X:", x)
print("Y:", y)
plt.grid()
plt.plot(x, y, c="r")
plt.fill_between(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
area = np.trapz(y)
print("Area =", area)
```

```python
# № 4.3.4

import matplotlib.pyplot as plt
import csv

X1 = []
X2 = []
X3 = []
Y1 = []
Y2 = []
Y3 = []
with open('E:/ForQuickAccess/apple.txt', 'r') as datafile:
    plotting = csv.reader(datafile, delimiter=',')
    for ROWS in plotting:
        X1.append(str(ROWS[0]))
        Y1.append(int(ROWS[1]))
with open('E:/ForQuickAccess/google.txt', 'r') as datafile:
    plotting = csv.reader(datafile, delimiter=',')
    for ROWS in plotting:
        X2.append(str(ROWS[0]))
        Y2.append(int(ROWS[1]))
with open('E:/ForQuickAccess/microsoft.txt', 'r') as datafile:
    plotting = csv.reader(datafile, delimiter=',')
    for ROWS in plotting:
        X3.append(str(ROWS[0]))
        Y3.append(int(ROWS[1]))
plt.figure(figsize=(12, 7))
plt.plot(X1, Y1, color='y', label='apple')
plt.plot(X2, Y2, color='r', label='google')
plt.plot(X3, Y3, color='g', label='microsoft')
plt.title('The 2021 year')
plt.xlabel('Months')
plt.ylabel('Share price')
plt.legend(fontsize=12)
ax = plt.gca()
ax.axes.xaxis.set_ticklabels([])
plt.show()
```

```python
# № 4.3.5

import math as m

x = float(input("Enter the first number: "))
y = float(input("Enter the second number: "))
print("Select an action:", "\n", "x + y", "\n", "x - y", "\n", "x * y", "\n", "x / y",
     "\n", "exp(x + y)", "\n", "sin(x + y)", "\n", "cos(x + y)", "\n", "x ^ y")
action = input("Action: ")
if action == "+":
    print("Answer =", x + y)
elif action == "-":
    print("Answer =", x - y)
elif action == "*":
    print("Answer =", x * y)
elif action == "/":
    if y == 0:
        print("Error: the denominator cannot be 0")
    else:
        print("Answer =", x / y)
elif action == "exp":
    print("Answer =", m.exp(x + y)) 
elif action == "sin":
    print("Answer =", m.sin(x + y))
elif action == "cos":
    print("Answer =", m.cos(x + y))
elif action == "^":
    print("Answer =", x ** y)
else:
    print("Please try again")
```
