```python
# № 1.3.4

import numpy as np

array = np.zeros((7, 7))
array[:, [0, -1]] = 1
array[[0, -1], :] = 1
print(array)
```

```python
# № 1.3.5

import numpy as np

array = np.array([2, 1, 5, 3, 7, 4, 6, 8, 9])
array = np.sort(array)[::-1]
print(array)
```

```python
# № 1.3.6

import numpy as np

array = np.random.random((3, 5))
print("Array:", "\n", array)
shape = np.shape(array)
size = np.size(array)
dimension = np.ndim(array)
print("Shape:", shape)
print("Size:", size)
print("Dimension:", dimension)
```

```python
# № 2.3.1

import pandas as pd
import math as m

print("Enter coordinates of the first point:")
x1, y1, = map(int, input().split())
print("Enter coordinates of the second point:")
x2, y2, = map(int, input().split())
x = pd.Series({'a': x1, 'b': x2})
y = pd.Series({'a': y1, 'b': y2})
distance = m.sqrt((x['a'] - x['b']) ** 2 + (y['a'] - y['b']) ** 2)
print("Euclidean distance =", distance)
```

```python
# № 3.3.2

import pandas as pd
from sklearn import preprocessing

url = 'https://raw.githubusercontent.com/akmand/datasets/master/iris.csv'
dataframe = pd.read_csv(url)
dataframe[["sepal_length_cm"]] = preprocessing.MinMaxScaler().fit_transform(dataframe[["sepal_length_cm"]])
dataframe[["sepal_width_cm"]] = preprocessing.StandardScaler().fit_transform(dataframe[["sepal_width_cm"]])
dataframe
```
